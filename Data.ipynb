```python
for house in houses_all:
    try:
        details = house.find('div',{'class':'col-md-12'}).find('div',{'class':'wp-block property list'}).find('div',{'class':'wp-block-footer'}).find('ul',{'class':'aux-info'}).find_all('li')
        info = house.find('div',{'class':'col-md-12'}).find('div',{'class':'wp-block property list'}).find('div',{'class':'wp-block-body'}).find('div',{'class':'wp-block-content clearfix text-xs-left text-sm-left'})
    except AttributeError:
        # Handle the case where a field is missing
        continue
```

```python
def housescrapping (house):
    house_Inf = {}
    info = house.find('div',{'class':'col-md-12'}).find('div',{'class':'wp-block property list'}).find('div',{'class':'wp-block-body'})
    house_Inf['Location'] = info.find('address',{'class':'voffset-bottom-10'}).strong.text.strip().split(',')[0].replace("\xa0", " ")
    house_Inf['Price'] = info.find('span',{'class':'pull-sm-left'}).text.strip()[4:]
    detail = house.find('div',{'class':'col-md-12'}).find('div',{'class':'wp-block property list'}).find('div',{'class':'wp-block-footer'}).find('ul',{'class':'aux-info'}).find_all('li')
    if len(detail)>0:
        if detail[0].span is not None:
            house_Inf['Bedrooms'] = detail[0].span.text
        else:
            house_Inf['Bedrooms'] = 'Not available'
    else:
        house_Inf['Bedrooms'] = 'Not available'
    if len(detail)>1:
        if detail[1].span is not None:
            house_Inf['Bathrooms'] = detail[1].span.text
        else:
            house_Inf['Bathrooms'] = 'Not available'
    else:
        house_Inf['Bathrooms'] = 'Not available'
    if len(detail)>2:
        if detail[2].span is not None:
            house_Inf['Toilets'] = detail[2].span.text
        else:
            house_Inf['Toilets'] = 'Not available'
    else:
        house_Inf['Toilets'] = 'Not available'
    if len(detail)>3:
        if detail[3].span is not None:
            house_Inf['Parking'] = detail[3].span.text
        else:
            house_Inf['Parking'] = 'Not available'
    else:
        house_Inf['Parking'] = 'Not available'
    return house_Inf
```

```python
housedict = [housescrapping (house) for house in houses_all]
data = pd.DataFrame(housedict)
data
```

```python
urls = []
for i in range(1,17):
    if i==1:
        url = "https://kenyapropertycentre.com/for-sale/houses/nairobi/showtype"
    else:
        url = f"https://kenyapropertycentre.com/for-sale/houses/nairobi/showtype?page={i}"
    urls.append(url)
urls
```

```python
def get_20_houses(url):
    htmlPage = requests.get(url)
    soup = BeautifulSoup(htmlPage.content,'html.parser')
    raw = soup.find_all('div',{'itemprop':'itemListElement'})
    to_dicts = [housescrapping (house) for house in raw]
    return to_dicts
```

```python
all_dict = []
for url in urls:
    all_dict.extend(get_20_houses(url))
print(len(all_dict))
all_dict
```

```python
data = pd.DataFrame(all_dict)
data
```
