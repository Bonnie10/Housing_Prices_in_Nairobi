```python
# Importing the neccessary libraries
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.the-star.co.ke/classifieds/house-apartment-for-sale/house--nairobi'
htmlPage = requests.get(url = url)
Soup = BeautifulSoup(htmlPage.content, 'html.parser')
```

```python
print(len(Soup.find_all('section',{'class':'product product-list-more product-no-premium image-left product_0'})))
Houses = Soup.find_all('section',{'class':'product product-list-more product-no-premium image-left product_0'})
```

```python
FirstHouse = Houses[0]
```

```python
FirstLocation = FirstHouse.find('div', {'class':"geo-meta-section"}).find('p',{'class':"product-top-meta"}).text
```

```python
FirstPrice = FirstHouse.find('div', {'class':"product-price"}).text.replace("KSh ", "")
```

```python
FirstBedrooms = FirstHouse.find('ul', {'class':"product-right-meta"}).find_all('li')[0].span.text.replace('bedrooms', '')
```

```python
FirstBathrooms = FirstHouse.find('ul', {'class':"product-right-meta"}).find_all('li')[1].span.text.replace('bathrooms', '')
```

```python
def HouseScrapping(house):
    infor = {}
    infor['Location'] = house.find('div', {'class':"geo-meta-section"}).find('p',{'class':"product-top-meta"}).text
    infor['Price'] = house.find('div', {'class':"product-price"}).text.replace("KSh ", "")
    infor['Bedrooms'] = house.find('ul', {'class':"product-right-meta"}).find_all('li')[0].span.text.replace('bedrooms', '')
    infor['Bathrooms'] = house.find('ul', {'class':"product-right-meta"}).find_all('li')[1].span.text.replace('bathrooms', '')
    return infor
```

```python
HousesDict = [HouseScrapping(house) for house in Houses]
data = pd.DataFrame(HousesDict)
data
```

```python
urls = ['https://www.the-star.co.ke/classifieds/house-apartment-for-sale/house--nairobi?p={}&s=default'.format (i) for i in range (1,1795)]
urls
```

```python
def get24houses(url):
    htmlPage = requests.get(url)
    soup = BeautifulSoup(htmlPage.content,'html.parser')
    raw = Soup.find_all('section',{'class':'product product-list-more product-no-premium image-left product_0'})
    to_dicts = [HouseScrapping(house) for house in raw]
    return to_dicts
    
```

```python
all_dict = []
for url in urls:
    all_dict.extend(get24houses(url))
print(len(all_dict))
all_dict
```

```python
data = pd.DataFrame(all_dict)
data
```

```python
data.to_csv('stardata.csv', index=False)
```
